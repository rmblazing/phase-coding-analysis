### First, plot example spike phase histograms and confidence intervals generated from shuffled spikes for each cell

```python
# for each cell, get the percentage of spikes exceeding the null confidence interval 
total_percent_spikes_exceeding_expt = shiva_data['total_percent_spikes_exceeding_expt']
all_percent_spikes_exceeding_cat = np.array([cell for expt in total_percent_spikes_exceeding_expt for cell in expt])
```

```python
# define a function to find the cell index with the value closest to the median - we will use this to plot the median example of phase locking
def find_nearest_idx(array, target):
    # first, sort the input array in ascending order
    sorted_array = np.sort(array)
    i = 1
    best_diff_idx = 0
    # find the difference between the fist value and the target value
    current_diff = abs(sorted_array[0]-target)
    while i < len(sorted_array):
        # if the difference between the updated value and the target is less than the current value and the target, save the updated value and update the current difference 
        if abs(sorted_array[i]-target)<current_diff:
            best_diff_idx = i
            current_diff = abs(sorted_array[i]-target)
        i+=1
    # get the value of the true index, rather than the sorted index
    true_idx = np.argsort(array)[best_diff_idx]
    return true_idx
```

```python
median_idx = find_nearest_idx(all_percent_spikes_exceeding_cat, np.median(all_percent_spikes_exceeding_cat))
```

```python
# Plot the median phase-locked cell 
cell = median_idx
bins = np.arange(0,np.radians(360)+np.radians(9), np.radians(9))
fig, ax = plt.subplots(1, 1, subplot_kw=dict(projection='polar'))
ax.bar(bins[:-1],all_phase_hist_cat[cell,:], align = 'edge', width = np.diff(bins), color = [.7,.7,.7])
ax.step(bins[:-1],all_radius_low_cat[cell,:], 'k', where="post",linewidth=1, color = 'k')
ax.step(bins[:-1],all_radius_high_cat[cell,:], 'k',where="post",linewidth=1, color = 'k')
plt.savefig(r'S:\All_Staff\robin\Paper\phase_coding_manuscript\figures\fig5\supplement1\median_phase_locked_cell.pdf', bbox_inches = 'tight')
```

### Plot percent of cells with spontaneous phase-locking

```python
# Load the phase-locking data. For each experiment, each cell is labeled with 1 (spontaneously locked, at least 5% of spikes exceeding the C.I.) or 0 (not significantly phase-locked)
spontaneous_phase_locked_expt = shiva_data['spontaneous_phase_locked_expt']
```

```python
all_percent_locked = []
for n,expt in enumerate(spontaneous_phase_locked_expt):
    percent_locked = (len(expt[expt>0])/len(expt))*100
    all_percent_locked.append(percent_locked)
```

```python
plt.plot(np.zeros(len(all_percent_locked)), all_percent_locked, 'kx', alpha = .4)

phase_locked_err = np.mean(all_percent_locked) -st.t.interval(alpha=0.95, df=len(all_percent_locked)-1, loc=np.mean(all_percent_locked), scale=st.sem(all_percent_locked))[0]

plt.errorbar(0, np.mean(np.array(all_percent_locked)), yerr = phase_locked_err, fmt = 'o', capsize = 3, elinewidth = 1, capthick = 1, linewidth = .5, color= 'k')


plt.ylim(0,100)
plt.bar(0, np.mean(all_percent_locked), width = .5, alpha = .2)
plt.xlim(-1,1)
plt.ylabel('% phase locked')
plt.xticks([])
PGanalysis.axis_fixer(ratio = 1, size = 20)
plt.savefig(r'S:\All_Staff\robin\Paper\phase_coding_manuscript\figures\fig5\supplement1\percent_cells_phase_locked_by_expt.pdf', bbox_inches = 'tight')
```
