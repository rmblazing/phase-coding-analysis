```python
def process_phase_data(expt_ID, mtrigger_file, expt_metadata_path, trial_type_psth_array_path, trial_type_raster_array_path, resp_phase_path, expt_category = 'contra', expt_type = 1):
    
    phase_stats_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig2\phase_stats'
    PCA_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig2\PCA'
    population_analysis_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig2\population_analysis_data'
    phase_tuning_curve_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig2\phase_tuning'
    spike_count_decoding_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig2\spike_count_decoding'
        
    process = phase_processor.Phase_data_analysis(mtrigger_file, expt_metadata_path, trial_type_psth_array_path, trial_type_raster_array_path, resp_phase_path, expt_type = expt_type)
    
    phase_stats_path = os.path.join(phase_stats_path, expt_category, expt_ID + '.npy')
    process.get_phase_stats(phase_stats_path)

    PCA_save_path = os.path.join(PCA_path, expt_category, expt_ID + '.npy')
    process.phase_trial_avg_PCA(PCA_save_path)

    pop_analysis_data_save_path = os.path.join(population_analysis_save_path, expt_category, expt_ID + '.npy')
    process.save_spike_count_decoder_data(pop_analysis_data_save_path)

    phase_tuning_curve_save_path = os.path.join(phase_tuning_curve_save_path, expt_category, expt_ID + '.npy')
    process.save_cell_stim_phase_tuning(phase_tuning_curve_save_path)

    decoding_save_path = os.path.join(spike_count_decoding_save_path, expt_category, expt_ID + '.npy')
    process.within_stim_phase_decoding_SVM(decoding_save_path)
```
