```python
import pickle as pkl
import numpy as np
import scipy
import matplotlib.pyplot as plt
import os
import phase_analysis
import preprocess_phase_data
import PGanalysis
```

```python
def preprocess_spiking_data(expt_ID, datafile, mtrigger_file, expt_metadata_path, expt_category = 'contra', expt_type = 0):
    
    raster_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig2\raster_data'
    psth_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig2\psth_data'
        
    raster_path = os.path.join(raster_save_path, expt_category, expt_ID + '.npy')
    psth_path = os.path.join(psth_save_path, expt_category, expt_ID + '.npy')
    preprocess_phase_data.save_phase_raster_psth(raster_path, psth_path, datafile, mtrigger_file, expt_metadata_path, expt_type = expt_type) 
```

```python
def preprocess_resp(expt_ID, resp_file):

    resp_phase_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig2\resp_data'
    
    resp_path = os.path.join(resp_phase_save_path, expt_ID + '.npy')
    resp_array, instantaneous_phase_corrected = preprocess_phase_data.save_instantaneous_phase(resp_path, resp_file)
    
    return resp_array, instantaneous_phase_corrected
```
