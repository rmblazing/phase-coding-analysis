```python
import pickle as pkl
import numpy as np
import scipy
import matplotlib.pyplot as plt
import os
import phase_analysis
import preprocess_phase_data
import PGanalysis
```

```python
def preprocess_spiking_data(expt_ID, datafile, mtrigger_file, expt_metadata_path, expt_category = 'main'):
    
    if expt_category == 'main':
        raster_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1\raster_data'
        psth_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1\psth_data'
        
    elif expt_category == 'supplement':
        raster_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\60micron_spots\raster_data'
        psth_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\60micron_spots\psth_data'
        
    elif expt_category == 'thy1':
        raster_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\thy1\raster_data'
        psth_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\thy1\psth_data'
        
    elif expt_category == 'bulb':
        raster_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\bulb\raster_data'
        psth_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\bulb\psth_data'
    
    elif expt_category == 'TeLC_PCx':
        raster_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig4\raster_data'
        psth_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig4\psth_data'
        
    elif expt_category == 'odor_pres':
        raster_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig6\raster_data'
        psth_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig6\psth_data'
        
    raster_path = os.path.join(raster_save_path, expt_ID + '.npy')
    psth_path = os.path.join(psth_save_path, expt_ID + '.npy')
    preprocess_phase_data.save_phase_raster_psth(raster_path, psth_path, datafile, mtrigger_file, expt_metadata_path) 
```

```python
def preprocess_resp(expt_ID, resp_file, expt_category = 'main'):
    
    if expt_category == 'main':
        resp_phase_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1\resp_data'
    elif expt_category == 'supplement':
         resp_phase_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\60micron_spots\resp_data'
    elif expt_category == 'thy1':
         resp_phase_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\thy1\resp_data'
    elif expt_category == 'bulb':
         resp_phase_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\bulb\resp_data'
    elif expt_category == 'TeLC_PCx':
         resp_phase_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig4\resp_data'
    elif expt_category == 'odor_pres':
         resp_phase_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig6\resp_data'
        
    resp_path = os.path.join(resp_phase_save_path, expt_ID + '.npy')
    resp_array, instantaneous_phase_corrected = preprocess_phase_data.save_instantaneous_phase(resp_path, resp_file)
    
    return resp_array, instantaneous_phase_corrected
```
