```python
def process_phase_data(expt_ID, mtrigger_file, expt_metadata_path, trial_type_psth_array_path, trial_type_raster_array_path, resp_phase_path, expt_category = 'main'):
    
    if expt_category == 'main':
        phase_stats_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1\phase_stats'
        PCA_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1\PCA'
        population_analysis_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1\population_analysis_data'
        phase_tuning_curve_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1\phase_tuning'
        spike_count_decoding_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1\spike_count_decoding'
        expt_type = 0
    
    elif expt_category == 'supplement':
        phase_stats_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\60micron_spots\phase_stats'
        PCA_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\60micron_spots\PCA'
        population_analysis_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\60micron_spots\population_analysis_data'
        phase_tuning_curve_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\60micron_spots\phase_tuning'
        spike_count_decoding_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\60micron_spots\spike_count_decoding'
        
    elif expt_category == 'thy1':
        phase_stats_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\thy1\phase_stats'
        PCA_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\thy1\PCA'
        population_analysis_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\thy1\population_analysis_data'
        phase_tuning_curve_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\thy1\phase_tuning'
        spike_count_decoding_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\thy1\spike_count_decoding'
        
    elif expt_category == 'bulb':
        phase_stats_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\bulb\phase_stats'
        PCA_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\bulb\PCA'
        population_analysis_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\bulb\population_analysis_data'
        phase_tuning_curve_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\bulb\phase_tuning'
        spike_count_decoding_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig1_supplement\bulb\spike_count_decoding'
           
    elif expt_category == 'TeLC_PCx':
        phase_stats_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig4\phase_stats'
        PCA_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig4\PCA'
        population_analysis_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig4\population_analysis_data'
        phase_tuning_curve_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig4\phase_tuning'
        spike_count_decoding_save_path = r'Z:\robin\Paper\phase_coding_manuscript\data\fig4\spike_count_decoding'
        
        
    process = phase_processor.Phase_data_analysis(mtrigger_file, expt_metadata_path, trial_type_psth_array_path, trial_type_raster_array_path, resp_phase_path)
    
    phase_stats_path = os.path.join(phase_stats_path, expt_ID + '.npy')
    process.get_phase_stats(phase_stats_path)

    PCA_save_path = os.path.join(PCA_path, expt_ID + '.npy')
    process.phase_trial_avg_PCA(PCA_save_path)

    pop_analysis_data_save_path = os.path.join(population_analysis_save_path, expt_ID + '.npy')
    process.save_spike_count_decoder_data(pop_analysis_data_save_path)

    phase_tuning_curve_save_path = os.path.join(phase_tuning_curve_save_path, expt_ID + '.npy')
    process.save_cell_stim_phase_tuning(phase_tuning_curve_save_path)

    decoding_save_path = os.path.join(spike_count_decoding_save_path, expt_ID + '.npy')
    process.within_stim_phase_decoding_SVM(decoding_save_path)
```
